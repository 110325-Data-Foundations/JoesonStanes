# Arrays and ufuncs(Universal functions) operations you can perform on arrays

In numpy we work with arrays in any dimention 
Ufuncs are built in functions that let us do math with those arrays(square every element of this array, add two arrays together,multiply two arrays together etc..Think back to matrix multiplication)


We can create these ones but generally these are built in ones  

```python
import numpy as np
```

```python
import numpy as np
arr1=np.array([1,2,3])
arr2=np.array([4,5,6])
res=np.multiply(arr1,arr2)
print(res)
```
