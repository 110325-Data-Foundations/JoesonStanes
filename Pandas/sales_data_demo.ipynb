## Basic Selection filtering

```python
#Selecting a column and everything in it
product_category = sales_df["product_category"]
#baisc filtering with booleans 
electronics_sales = sales_df[sales_df["product_category"]=="Electronics"]
electronics_sales.head()
```

```python
## Basic Selection filtering
not_sales_electronics = sales_df[sales_df["product_category"]!="Electronics"]
not_sales_electronics.head(20)
```

## Section 3: Feature Engineering + Data Cleaning

```python

#Create a new dataframe with only high value sales
sales_df["total_sales"] = sales_df["unit_price"] * sales_df["quantity"]
high_value_df = sales_df[sales_df["total_sales"]>100]
high_value_df.sort_values("total_sales", ascending=False)
high_value_df.head(10)
```

```python
print(sales_df.shape[1])
sales_df.columns
```

```python
#yesterday we saw that we camn throuw out rows and values with missing data

# we probably dont want to skew any analysisi operations or visualizations we do thath badly by neterimng zeros 
# we can fill these values with a mean if its numberaiacl

sales_df['unit_price']=sales_df['unit_price'].fillna(sales_df['unit_price'].mean())
sales_df.head()
```

```python
# We can  also handle non numerical columns by filling them with the default values
sales_df['region']=sales_df['region'].fillna('Unknown')
sales_df.head(20
              )
```

## DATA AGGREGATIONS AND GROUP BY 

```python
#we want the total salws by product categories 
sales= sales_df.groupby("product_category")["total_sales"].sum()
#Average transaction value by region
avg_sales_region = sales_df.groupby("region")["total_sales"].mean()
display(sales)
```

```python
##We can run multiple aggregations using .agg()
agg_sales = sales_df.groupby("product_category").agg({
    'total_sales':['sum','mean','max','min','count'],
    'quantity':['sum','mean']})
display(agg_sales)
```

## Visulaisations

We'll often find ourselves creating just a series if disjoint graphs and plots, but a coherent dashboard with severa; subplots within it

```python
#Lets reate a sinple dsashboard with multiple plots
import matplotlib.pyplot as plt

fig,axes = plt.subplots(2,2, figsize=(15,6))
agg_sales.plot(kind='bar', ax=axes[0,0], title='Total Sales by Product Category')
#sales trend per month 
#Deriving month from order date 
#dt lets us extract things from the date time data in a cell
sales_df['order_date'] = pd.to_datetime(sales_df['order_date'])
sales_df['month'] = sales_df['order_date'].dt.month
monthley_sales = sales_df.groupby('month')['total_sales'].sum()

monthley_sales.plot(kind='line', ax=axes[0,1], title='Monthly Sales Trend')
axes[0,1].set_xlabel('Month')
axes[0,1].set_ylabel('Total Sales')
```
